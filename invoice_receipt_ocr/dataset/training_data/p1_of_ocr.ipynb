```python

model = YOLO("yolov8n-cls.pt")

model.train(
    data="C:/Users/princ/OneDrive/Desktop/New folder (3)/dataset/training_data",
    epochs=20,
    imgsz=224,
    batch=8,
    workers=0
)
```

```python
import cv2

model = YOLO("runs/classify/train-4/weights/best.pt")


image_path = "C:/Users/princ/OneDrive/Desktop/New folder (3)/billofhotel.jpg"

img = cv2.imread(image_path)

# Check image loaded
if img is None:
    print("Image not found")
    exit()



results = model(img)

# First result
result = results[0]

class_id = result.probs.top1

# Class name
class_name = result.names[class_id]

# Confidence
confidence = result.probs.top1conf.item()



print("Prediction :", class_name)
print("Confidence :", round(confidence, 2))



text = f"{class_name} : {confidence:.2f}"

cv2.putText(
    img,
    text,
    (20, 50),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (0, 255, 0),
    2
)


cv2.namedWindow("Prediction", cv2.WINDOW_NORMAL)

cv2.resizeWindow("Prediction", 800, 600)


cv2.imshow("Prediction", img)

cv2.waitKey(0)

cv2.destroyAllWindows()
```
