```python
#only video capture with webcam
from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO("yolov8n.pt")

# Open webcam
cap = cv2.VideoCapture(0)



# Create smaller resizable window
cv2.namedWindow("Live Detection", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Live Detection", 800, 600)

while True:

    # Read frame from webcam
    ret, frame = cap.read()

    if not ret:
        print("Camera not working")
        break

    # Run YOLO detection
    results = model(frame)

    # Draw detections
    annotated_frame = results[0].plot()

    # Show result
    cv2.imshow("Live Detection", annotated_frame)

    # Press Q to quit
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# Release camera
cap.release()

# Close all windows
cv2.destroyAllWindows()
```

```python
# video capture with save in folder 
from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO("yolov8n.pt")

# Open webcam
cap = cv2.VideoCapture(0)

# Get webcam width and height
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

# Define codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')

# Output video file name
out = cv2.VideoWriter(
    "output_detection.avi",
    fourcc,
    20.0,
    (frame_width, frame_height)
)

# Create resizable window
cv2.namedWindow("Live Detection", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Live Detection", 800, 600)

while True:

    # Read frame
    ret, frame = cap.read()

    if not ret:
        print("Camera not working")
        break

    # YOLO Detection
    results = model(frame)

    # Draw detections
    annotated_frame = results[0].plot()

    # Save frame into video
    out.write(annotated_frame)

    # Show frame
    cv2.imshow("Live Detection", annotated_frame)

    # Press Q to quit
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# Release everything
cap.release()
out.release()

# Close windows
cv2.destroyAllWindows()

print("Video saved successfully")
```
